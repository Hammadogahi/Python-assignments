```python
todo = []
run = False
delete = []
instructions = print('''
Instructions
new- Add new todo
delete all -delete todo
delete -delete todo
show -show the list of todo
exit- exit todo
''')


while run != True:
    while run == False:
        command = input('Please enter the command\n').lower()
        run == True
        break
    if command.lower() == 'new':
        add_todo = input('Enter the todo').lower()
        todo.append(add_todo)
        run = False
    elif command.lower() == 'delete all':
        todo.clear()
        print('Deleted Successfully')
        run = False
    elif command.lower() == 'show':
        for index, name in enumerate(todo, start=1):
            print(" {}: {}".format(index, name))
    elif command.lower() == 'delete':
        delete_todo = input('Please write the todo that you want to delete').lower()
        delete.append(delete_todo)
        for name in delete:
            todo.remove(name)
            print('deleted Successfully\n Type show to view list')
            run = False
    elif command == 'exit':
        print('Program ended')
        break
    else:
        print('Please enter a valid Command')
        
```
